```python
engine(bits=10)
x = integer()
assert x <= 1
while satisfy():
    print(x)
```

```python
engine(bits=10)
x = integer()
y = integer()
z = integer()
assert x + y == z
assert x <= 10
assert y <= 5
assert z <= 7
while satisfy():
    print(x, y, x + y == z)
```

```python
engine(bits=10)
x = integer()
y = integer()
z = integer()
assert x ** 2 + y ** 2 == z ** 2
assert 0 < x <= 100
assert 0 < y <= 50
assert 0 < z <= 70
while satisfy():
    print(x, y, x ** 2 + y ** 2 == z ** 2)
```

```python
engine(bits=10)
x = integer()
y = integer()
z = integer()
assert x * y == z
assert 0 < x <= 10
assert 0 < y <= 5
assert 0 < z <= 7
while satisfy():
    print(x, y, x * y == z)
```

```python
n = 5
engine(bits=10)
xs = [integer() for _ in range(n)]
assert sum(xs) == n
while satisfy():
    print(xs, sum(xs) == n)
```

```python
n = 5
engine(bits=10)
xs = vector(size=n)
assert sum(xs) == n
while satisfy():
    print(xs, sum(xs) == n)
```

```python
n, m = 3, 2
engine(bits=10)
xs = matrix(dimensions=(n, m))
for i in range(n):
    for j in range(m):
        xs[i][j] <= 1
while satisfy():
    print(xs)
```

```python
n, m = 3, 2
engine(bits=10)
xs = matrix(dimensions=(n, m))
all_binaries(xs)
while satisfy():
    print(xs)
```

```python
n, m = 3, 3
engine(bits=10)
xs = matrix(dimensions=(n, m))
all_binaries(xs)
for i in range(n):
    assert sum(xs[i][j] for j in range(m)) == 1
for j in range(m):
    assert sum(xs[i][j] for i in range(n)) == 1
while satisfy():
    print(xs)
```

```python
n, m = 3, 3
engine(bits=10)
xs = matrix(dimensions=(n, m))
for i in range(n):
    assert sum(xs[i][j] for j in range(m)) == n
for j in range(m):
    assert sum(xs[i][j] for i in range(n)) == m
while satisfy():
    print(xs)
```
