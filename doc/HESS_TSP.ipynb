```
import numpy as np
import matplotlib.pyplot as plt
import peqnp
```

```
def oracle(seq):
  return sum(np.linalg.norm(data[seq[i - 1]] - data[seq[i]]) for i in range(len(seq)))
```

```
n = 100
data = np.random.logistic(size=(n, 2))
xs, ys = zip(*data)
plt.plot(xs, ys, 'ro')
plt.show()
```

```
seq = peqnp.hess_sequence(n, oracle)
```

```
xs, ys = zip(*[data[i] for i in seq + [seq[0]]])
plt.plot(xs, ys, 'k-')
plt.plot(xs, ys, 'ro')
plt.show()
```
